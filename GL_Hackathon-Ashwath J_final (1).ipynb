# SVM

# DT

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score
from sklearn import tree
dtree = DecisionTreeClassifier(criterion='entropy',random_state=1,max_depth = 6)
dtree.fit(X,y)
```

# RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 60,random_state=1,max_features = 10,criterion='entropy')
rfc.fit(X,y)
```

# ADABOOST

```python
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier(n_estimators = 120 ,random_state = 1,algorithm='SAMME')
abc.fit(X,y)
```

# STACKING MODEL

```python
from sklearn.ensemble import StackingClassifier
from sklearn.neural_network import MLPClassifier
est = [('rfc',rfc),('abc',abc),('dt',dtree)]
stack_model = StackingClassifier(estimators = est,final_estimator = MLPClassifier(random_state = 1))
stack_model.fit(X,y)
```
